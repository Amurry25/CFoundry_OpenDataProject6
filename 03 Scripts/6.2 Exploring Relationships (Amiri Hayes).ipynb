# 0.3 Matrix Heatmap

```python
# Create a subset excluding the "Country", "Region" and "Year" columns
sub = df_clean2[['Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Family (Social Support)', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity']]
```

```python
# Create a subplot w/ matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

#### Correlation between Happiness Score & _____
- Economy (GDP per Capita) (0.79) = Strong correlation
- Health (Life Expectancy) (0.74) = Strong correlation
- Family (Social Support) (0.65) = Strong correlation
- Generosity (0.14) = Weak correlation

# 0.6 Categorical plot

```python
df_clean2.head()
```

```python
# Use a histogram to visualize the distribution of the dependent variable = 'Happiness Score'
sns.histplot(df_clean2['Happiness Score'], bins = 5, kde = True)
```

```python
# Creating histogram to visualize 'Happiness Score' distribution
sns.histplot(df_clean2['Happiness Score'], bins = 50, kde = True)
```

```python
#Creating a categorical variable that splits the "Happiness Score" column into categories
df_clean2.loc[df_clean2['Happiness Score'] < 4, 'Happiness Rate'] = 'Low Rate'
```

```python
df_clean2.loc[(df_clean2['Happiness Score'] >= 4) & (df_clean2['Happiness Score'] < 6), 'Happiness Rate'] = 'Middle Rate'
```

```python
df_clean2.loc[df_clean2['Happiness Score'] >= 6, 'Happiness Rate'] = 'High Rate'
```

```python
df_clean2['Happiness Rate'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the "Happiness Score" and other variables
sns.set(style="ticks")
sns.scatterplot(x = 'Economy (GDP per Capita)', y = 'Happiness Score', hue = 'Year', data = df_clean2)
```

```python
sns.set(style="ticks")
sns.scatterplot(x = 'Health (Life Expectancy)', y = 'Happiness Score', hue = 'Year', data = df_clean2)
```

```python
sns.set(style="ticks")
sns.scatterplot(x = 'Economy (GDP per Capita)', y='Happiness Score', hue='Happiness Rate', data=df_clean2)
```
