### 0.3 Elbow Technique 

```python
num_cl = range(1, 10) # Defines the range of potential clusters in the data.
kmeans = [KMeans(n_clusters=i) for i in num_cl] # Defines k-means clusters in the range assigned above.
```

```python
score = [kmeans[i].fit(df_clean3_no_categorical).score(df_clean3_no_categorical) for i in range(len(kmeans))] # Creates a score that represents 
# a rate of variation for the given cluster option.

score
```

```python
# Plot the elbow curve using PyLab.
pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

#####  In the example above, there's a large jump from two to three on the x-axis, but after that, the curve straightens out. This means that the optimal count for your clusters is three.

### 0.4 k-means clustering

```python
# Create the k-means object.
kmeans = KMeans(n_clusters = 3, n_jobs = -1) 
```

```python
# Fit the k-means object to the data.
kmeans.fit(df_clean3_no_categorical)
```

```python
df_clean3_no_categorical['clusters'] = kmeans.fit_predict(df_clean3_no_categorical)
```

```python
df_clean3_no_categorical.head()
```

```python
df_clean3_no_categorical['clusters'].value_counts()
```

```python
# Plot the clusters for the "Economy (GDP per Capita)" and "Happiness Score" variables.
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df_clean3_no_categorical['Economy (GDP per Capita)'], y=df_clean3_no_categorical['Happiness Score'], hue=kmeans.labels_, s=100) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels. 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm.
# `s` represents the size of the points you want to see in the plot.

ax.grid(False) # This removes the grid from the background.
plt.xlabel('Economy (GDP per Capita)') # Label x-axis.
plt.ylabel('Happiness Score') # Label y-axis.
plt.show()
```

```python
# Plot the clusters for the "Family (Social Support)" and "Happiness Score" variables.
plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df_clean3_no_categorical['Family (Social Support)'], y=df_clean3_no_categorical['Happiness Score'], hue=kmeans.labels_, s=100)

ax.grid(False) 
plt.xlabel('Family (Social Support)') 
plt.ylabel('Happiness Score') 
plt.show()
```

```python
df_clean3_no_categorical.loc[df_clean3_no_categorical['clusters'] == 2, 'cluster'] = 'dark purple'
df_clean3_no_categorical.loc[df_clean3_no_categorical['clusters'] == 1, 'cluster'] = 'purple'
df_clean3_no_categorical.loc[df_clean3_no_categorical['clusters'] == 0, 'cluster'] = 'pink'
```

Both the clusters (2) shown above appear to for the most part to have linear relationships as both variables correlate with one another. 

```python
df_clean3_no_categorical.groupby('cluster').agg({'Freedom':['mean', 'median'], 
                         'Family (Social Support)':['mean', 'median'], 
                         'Economy (GDP per Capita)':['mean', 'median'],
                          'Happiness Score':['mean', 'median']})
```
